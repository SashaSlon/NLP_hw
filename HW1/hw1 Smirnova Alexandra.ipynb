```python
import os
import nltk
from nltk import sent_tokenize
import re
import pandas as pd
```

Книга: Harry_Potter_ru.txt

Идеальное разбиение предложений: gold_sentences.txt

```python
with open(r"Harry_Potter_ru.txt", 'r', encoding = 'utf-8') as text:
    tex = text.read()
with open('gold_sentences.txt', 'r', encoding = 'utf-8') as g:
    hp = g.readlines()
    hp = [s[:-1] for s in hp] #убираю \n с конца предложений
```

```python
hp
```

Разбиваю кусок на предложения регуляркой. 

```python
def splitterizator(x):
    x = re.sub(r'([\.!\?]) ([А-ЯЁ])', r'\1\n\2', x)
    x = re.split(r'\n', x)
    return x
```

```python
#склеиваю идеально разбиенный отрывок в сплошной текст
compare = ' '.join(hp)
compare = re.sub('\n', '', compare)
compare
```

```python
ken = splitterizator(compare)
```

```python
ken #текст, разбитый на предложения регуляркой
```

С помощью двух метрик оцениваю качество разбиения

```python
print('Gold length - ', len(hp))
print('My split - ', len(ken))
print('Percentage of overlaps', str(len(set(hp) & set(ken)) / len(set(hp) | set(ken))))
```

```python
tp = 0
fp = 0
fn = 0

for sent in hp:
    if len(splitterizator(sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(hp)-1):
    sent1, sent2 = hp[i], hp[i+1]
    sent = ' '.join([sent1, sent2])
    if len(splitterizator(sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision)
print('Recall - ', recall)
print('F1 - ', f1)
```

#### Обучаю PunktSentenceTokenizer на всем тексте

```python
from pprint import pprint
from nltk.tokenize.punkt import PunktSentenceTokenizer, PunktTrainer
```

```python
trainer = PunktTrainer()
trainer.INCLUDE_ALL_COLLOCS = True
trainer.train(tex)

tokenizer = PunktSentenceTokenizer(trainer.get_params())
```

Разбиваю выбранный кусок на предложения через PunktSentenceTokenizer и оцениваю качество.

```python
bob = tokenizer.tokenize(compare)
bob
```

```python
print('Gold length - ', len(hp))
print('My split - ', len(bob))
print('Percentage of overlaps', str(len(set(hp) & set(bob)) / len(set(hp) | set(bob))))
```

```python
tp = 0
fp = 0
fn = 0

for sent in hp:
    if len(tokenizer.tokenize(sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(hp)-1):
    sent1, sent2 = hp[i], hp[i+1]
    sent = ' '.join([sent1, sent2])
    if len(tokenizer.tokenize(sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision)
print('Recall - ', recall)
print('F1 - ', f1)
```
